#### 2.1. List all language names and number of related articles in the order they appear in wikipedia.org.

````python
# This is the url you will scrape in this exercise
url = 'https://www.wikipedia.org/'
````

````python
# your code here
response = requests.get(url)
response.status_code

soup = BeautifulSoup(response.content)
soup

languages1 = soup.find_all("strong")
languages1
languages = [i.text for i in languages1]
languages = languages[1:-1]
languages

articles1 = soup.find_all("bdi", attrs={"dir":"ltr"})
articles1
articles = [i.text.replace("\xa0","") for i in articles1]
articles = articles[:10]
articles

lang_art = list(zip(languages, articles))
lang_art
````

#### 3. Display IMDB's top 250 data (movie name, initial release, director name and stars) as a pandas dataframe.
Hint: If you hover over the title of the movie, you should see the director's name. Can you find where it's stored in the html?

````python
# This is the url you will scrape in this exercise 
url = 'https://www.imdb.com/chart/top'
````

````python
# your code here
response = requests.get(url)
response.status_code

soup = BeautifulSoup(response.content)
soup

table = soup.find_all("td", attrs={"class":"titleColumn"})
table

movie = [i.find("a").text for i in table]
movie

director = [i.find("a").get("title") for i in table]
director

release = [i.find("span").text for i in table]
release

rating = soup.find_all("td", attrs={"class":"ratingColumn imdbRating"})
stars = [i.find("strong").text for i in rating]
stars

movies_final = pd.DataFrame(list(zip(movie, release, director, stars)), columns=["Movie_Name", "Initial_Release", "Director_Name", "Rating"])
movies_final
````

#### 3.1. Display the movie name, year and a brief summary of the top 10 random movies (IMDB) as a pandas dataframe.

````python
#This is the url you will scrape in this exercise
url = 'https://www.imdb.com/list/ls009796553/'
````

````python
# your code here
response = requests.get(url)
response.status_code

soup = BeautifulSoup(response.content)
soup

movies_info = soup.find_all("div", attrs={"class":"lister-item-content"})
movie_title = [i.find("a").text for i in movies_info]
movie_title

year = movies_info[3].find("span")
year = [i.find_all("span", attrs={"class":"lister-item-year text-muted unbold"})[0].text for i in movies_info]
year

description = [i.find_all("p",attrs={"class":""})[0].text.strip() for i in movies_info]
description

top10 = pd.DataFrame(list(zip(movie_title, year, description)), columns=["Movie_Name", "Release_Year", "Summary"])
top10 = top10[:11]
top10
````
